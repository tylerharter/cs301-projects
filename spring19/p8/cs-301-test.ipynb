```python
#q1
def get_mapping(filename):
    reader = csv.reader(open(filename, encoding = "utf8"))
    mydict = {row[0] : row[1] for row in reader}
    return mydict

mapping1 = get_mapping("mapping.csv")
mapping1["nm0000233"]
```

```python
def get_movies(filename):
    '''
    Returns a list in which each element is a dictionary having all the movie's details
    :param filename: path to the movie csv file
    :return: returns list with the dictionaries
    '''
    reader = csv.reader(open(filename, encoding = "utf8"))
    dict_list = []
    for row in reader:
        dict1 = {
        "title"     : row[0],
        "year"     : int(row[1]),
        "rating"    : float(row[2]),
        "directors" : list(row[3].split(",")),
        "actors"    : list(row[4].split(",")),
        "genres"    : list(row[5].split(","))}
        dict_list.append(dict1)
    return dict_list

all_movies = get_movies("movies.csv")
print_dict(all_movies)
```

```python
def read_data(movie_filename, mapping_filename):    
    mapping_dict = get_mapping(mapping_filename)
    movie_dict = get_movies(movie_filename)
    movie_true_dict = movie_dict
    #Traverse the length of the list of dictionaries
    for i in range(len(movie_dict)):
        # For each dictionary access the actors section and find a mapping for each of the elements of the list contained in the key value
        for j in range(len(movie_dict[i]["actors"])):
            movie_true_dict[i]["actors"][j] = mapping_dict[movie_dict[i]["actors"][j]]

        for k in range(len(movie_dict[i]["directors"])):
            movie_true_dict[i]["directors"][k] = mapping_dict[movie_dict[i]["directors"][k]]

        movie_true_dict[i]["title"] = mapping_dict[movie_dict[i]["title"]]
    return movie_true_dict


movie_dict = read_data("movies.csv", "mapping.csv")
print_dict(movie_dict)
```

```python
movies_list = []
actors_list = []
directors_list = []
genres_list = []

def list_generator(movie_dict):
    global movies_list
    global actors_list
    global directors_list
    global genres_list
    
    for i in range(len(movie_dict)):
        
        movies_list.append(movie_dict[i]["title"])
        
        for j in range(len(movie_dict[i]["actors"])):
            actors_list.append(movie_dict[i]["actors"][j])
        
        for j in range(len(movie_dict[i]["directors"])):
            directors_list.append(movie_dict[i]["directors"][j])
        
        for j in range(len(movie_dict[i]["genres"])):
            genres_list.append(movie_dict[i]["genres"][j])
    
    genres_list = sorted(list(set(genres_list)))
    actors_list = sorted(list(set(actors_list)))
    directors_list = sorted(list(set(directors_list)))
    movies_list = sorted(list(set(movies_list)))
    print("Lists created !")

list_generator(movie_dict)
```

```python
#q9
def stats(movies_list, genres_list, directors_list, actors_list):
    stats_dict = {
            "num_movies"    : len(movies_list),
            "num_actors"    : len(actors_list),
            "num_directors" : len(directors_list),
            "num_genres"    : len(genres_list)
    }

    return stats_dict


stats_dict = stats(movies_list, genres_list, directors_list, actors_list)
stats_dict
```

```python
def career_span(movie_dict,use_case, start=0,end=5):
    score={}
    career_score = {}
    for movie in movie_dict:
        for item in movie[use_case]:
            if item not in score.keys():
                score[item]=[int(movie["year"])]
            else:
                score[item].append(int(movie["year"]))
    
    for item in score:
        career_score[item] = max(score[item]) - min(score[item])
        
    sorted_x = dictionary_sort(career_score)
    return sorted_x[start:end]

career_span(movie_dict, "directors", 0, 15)
```

```python
def movies_per_year(movie_dict, year):
    year_wise_score = {}
    for movie in movie_dict:
        if movie["year"] not in year_wise_score.keys():
            year_wise_score[movie["year"]] = 1
        else:
            year_wise_score[movie["year"]] += 1
            
    #sorted_x = dictionary_sort(year_wise_score)         
    return year_wise_score[year]

```

```python
def genre_list(movie_dict,category):
    genre={}
    for movie in movie_dict:
        for each_genre in movie["genres"]:
            if each_genre not in genre.keys():
                genre[each_genre]=[movie["title"]]
            else:
                (genre[each_genre]).append(movie["title"])
          
    return sorted(genre[category])

#genre_list(movie_dict, "Horror")
```

```python
def year_ratings(movie_dict, start = 0, end = 5):
    year_wise_ratings = {}
    year_wise_score = {} 
    for movie in movie_dict:
        if movie["year"] not in year_wise_ratings.keys():
            year_wise_ratings[movie["year"]] = [movie["rating"]]
        else:
            year_wise_ratings[movie["year"]].append(movie["rating"])
    
    for year in year_wise_ratings:
        year_wise_score[year] = round(sum(year_wise_ratings[year])/len(year_wise_ratings[year]),2)
    
    sorted_x = dictionary_sort(year_wise_score)            
    return sorted_x[start:end]
        
```
